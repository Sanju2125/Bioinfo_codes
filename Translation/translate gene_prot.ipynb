```python
#Creating a dictionary of genetic code
a1=open("genetic_code.txt","r");
a2=a1.readlines();
embs={};
for i in range(11,len(a2[0])):
    aa=a2[0][i];
    codon=a2[2][i]+a2[3][i]+a2[4][i];
    embs[codon]=aa;
print(embs);
```

```python
#Storing the fasta file
import re;
sq=open("collagen.fasta","r");
sa=sq.readlines();
seq={};
for j in sq:
    j=j.rstrip("\n");
    xyz=re.match(">",j);
    if(xyz):
        header=j;
        seq[j]="";
    else:
        seq[header]=seq[header]+j;
    print(seq);
```

```python
fasta=seq.keys()
for j in fasta:
    dna=seq[j]
    x=0
    while(x<len(dna)):
        print(dna[x]+dna[x+1]+ dna[x+2])
        x=x+3;
```

```python
fasta=seq.keys()
for j in fasta:
    dna=seq[j]
    x=0
    aa=""
    while(x<len(dna)-2):
        myseq=dna[x]+dna[x+1]+ dna[x+2]
        if(myseq in embs.keys()):
            aa=aa+embs[myseq];
        x=x+3;
    print(j+dna +"\n" + "\n"+ "translated protein is below:\n", aa + "\n\n")
```

```python
#Translate
out=open("result.txt","w");

fasta=seq.keys()
for j in fasta:
    dna=seq[j]
    x=0
    aa=""
    while(x<len(dna)-2):
        myseq=dna[x]+dna[x+1]+ dna[x+2]
        if(myseq in embs.keys()):
            aa=aa+embs[myseq];
        x=x+3;
    out.write("\n"+j+"\n"+dna+"\n"+aa+"\n\n");
```
