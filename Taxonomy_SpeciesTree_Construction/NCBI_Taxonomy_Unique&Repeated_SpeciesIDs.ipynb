```python
import os

# Set the ETE_NCBI_TAXONOMY_DB environment variable
os.environ['ETE_NCBI_TAXONOMY_DB'] = r"D:/Documents/NCBI Taxonomy FTP Data/taxdump"

# Print the environment variable to verify
print("ETE_NCBI_TAXONOMY_DB set to:", os.environ['ETE_NCBI_TAXONOMY_DB'])

from ete3 import NCBITaxa

# Initialize NCBI Taxonomy with local database
ncbi = NCBITaxa()

# Load species names from a file
with open("C:/Users/DELL/Desktop/HGT Files/HGT TAXA 520.txt") as f:
    species_names = [line.strip() for line in f]

# Define a manual taxid mapping for known species
manual_taxid_mapping = {
    "Elaeis guineensis var. tenera": 51953,
    "Candidozyma haemuloni var. vulneris": 1231523,
    "uncultured marine thaumarchaeote SAT1000_52_H09": 1167203,
    "Helotiales sp. DMI_Dod_QoI": 2728926,
    "Kazachstania exigua": 34358,
    "Gibberella moniliformis":405761,
    # Add others as needed
}
# Get NCBI taxid for each species
species_taxids = []
for species in species_names:
    # First, check if the species is in the manual mapping
    if species in manual_taxid_mapping:
        species_taxids.append(manual_taxid_mapping[species])
    else:
        # If not in manual mapping, get the taxid from NCBI
        taxid = ncbi.get_name_translator([species])
        if species in taxid:
            species_taxids.append(taxid[species][0])
        else:
            print(f"No match found for {species}")

# Build the species tree if valid taxids were found
if species_taxids:
    species_tree = ncbi.get_topology(species_taxids)

    # Visualize the tree
    species_tree.show()

    # Save the tree to a file
    species_tree.write(format=1, outfile="species_tree.nwk")
    print("Species tree saved as 'species_tree.nwk'")
else:
    print("No valid taxids found. Check your species names.")
```

```python
# Read the species names from a text file
input_file = "C:/Users/DELL/Desktop/HGT Files/HGT TAXA 520.txt"  # Replace with your filename
unique_output_file = "C:/Users/DELL/Desktop/HGT Filesunique_species.txt"
repeated_output_file = "C:/Users/DELL/Desktop/HGT Filesrepeated_species.txt"

# Read the file and store species names in a list
with open(input_file, "r") as file:
    species = [line.strip() for line in file]

# Count occurrences of each species name
from collections import Counter
species_counts = Counter(species)

# Separate unique and repeated species
unique_species = [taxon for taxon, count in species_counts.items() if count == 1]
repeated_species = [taxon for taxon, count in species_counts.items() if count > 1]

# Print the results
print(f"Total species: {len(species)}")
print(f"Unique species: {len(unique_species)}")
print(f"Repeated species: {len(repeated_species)}")

# Write results to files
with open(unique_output_file, "w") as unique_file:
    unique_file.write("\n".join(unique_species))

with open(repeated_output_file, "w") as repeated_file:
    repeated_file.write("\n".join(repeated_species))

print(f"Unique species saved to: {unique_output_file}")
print(f"Repeated species saved to: {repeated_output_file}")
```
