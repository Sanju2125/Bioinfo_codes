```python
#Creating a dict within dict for a pdb file
import os;
os.chdir("D:\\EMBS\\Python")
abc=open("1BOM.pdb","r");
pqr=abc.readlines();
pro={};
for i in pqr:
    i=i.split();
    xyz=i[0];
    if(xyz=="ATOM"):
        aa=i[3];
        atom=i[-1];
        if aa in pro:
            if atom in pro[aa]:
                pro[aa][atom]+=1;
            else:
                pro[aa][atom]=1;
        else:
            pro[aa]={};
            pro[aa][atom]=1;
print(pro);            
        
```
